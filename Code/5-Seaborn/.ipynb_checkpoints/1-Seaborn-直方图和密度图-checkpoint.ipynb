# Seaborn-直方图和密度图

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
s1 = pd.Series(np.random.randn(1000))
```

```python
plt.hist(s1)
```

```python
s1.plot(kind='kde')
```

```python
sns.distplot(s1, rug=True)
```

```python
sns.kdeplot(s1)
```

```python
sns.rugplot(s1)
```
