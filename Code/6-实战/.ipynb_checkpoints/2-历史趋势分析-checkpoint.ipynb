```python
# 基本信息
import numpy as np
import pandas as pd

# 股票数据读取
import pandas_datareader as pdr

# 可视化
import matplotlib.pyplot as plt
import seaborn as sns

# time
from datetime import datetime
```

```python
start = datetime(2015,9,20)
alibaba = pdr.get_data_yahoo('BABA', start=start)
amazon = pdr.get_data_yahoo('AMZN', start=start)
```

```python
alibaba.head()
```

```python
alibaba['Adj Close'].plot(legend=True, label='Alibaba', title='Adj Close')
amazon['Adj Close'].plot(legend=True, label='Amazon')
```

```python
alibaba['Volume'].plot(legend=True, label='Alibaba', title='Volume')
amazon['Volume'].plot(legend=True, label='Amazon')
```

```python
alibaba['high-low'] = alibaba['High'] - alibaba['Low']
amazon['high-low'] = amazon['High'] - amazon['Low']
```

```python
alibaba['high-low'].plot(legend=True, label='Alibaba', title='high-low')
amazon['high-low'].plot(legend=True, label='Amazon')
```

```python
# daily returns
alibaba['daily return'] = alibaba['Adj Close'].pct_change()
amazon['daily return'] = amazon['Adj Close'].pct_change()
```

```python
alibaba['daily return'].plot(legend=True, label='Alibaba', title='daily return', figsize=(12, 4))
```

```python
alibaba['daily return'].plot(kind='hist')
```
